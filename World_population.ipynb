```python
import json, pygal
from pygal_maps_world.maps import World
from pygal_maps_world.maps import COUNTRIES
from pygal.style import LightColorizedStyle as LCS, RotateStyle as RS
```

```python
def get_country_code(country_name):
    for code,name in COUNTRIES.items():
        if name == country_name:
            return code
    return None
```

```python
filename = 'population_json.json'
with open(filename) as f:
    pop_data = json.load(f)
    
#print the 2010 population of each country.
cc_population = {}
for pop_dict in pop_data:
    if pop_dict['Year'] == 2016:
        country_name = pop_dict['Country Name']
        population = int(float(pop_dict['Value']))
        code = get_country_code(country_name)
        if code:
            cc_population[code] = population
            
cc_pops1, cc_pops2, cc_pops3 = {}, {}, {}
for cc, pop in cc_population.items():
    if pop < 10000000:
        cc_pops1[cc] = pop
    elif pop < 1000000000:
        cc_pops2[cc] = pop
    else:
        cc_pops3[cc] = pop

wp_style = RS('#369', base_style=LCS)
wp = World(style=wp_style)
wp.title = 'World Populationin 2016, by Country'
wp.add('0 - 10m', cc_pops1)
wp.add('10m - 1bn', cc_pops2)
wp.add('>1bn', cc_pops3)
wp.render_to_file('world_population_colored.svg')
```
